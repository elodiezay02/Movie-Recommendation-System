```python
#Merge_data

meta = meta.merge(credit, on='id')
meta = meta.merge(keyword, on='id')
```

```python
#Lọc ra tất cả các giá trị 'id' trong small_dataset mà có trong links_small
#Hàm isin(): check 1 giá trị có nằm trong tập data hay ko. Nếu có thì giữ lại, nếu ko sẽ drop

small_dataset = meta[meta['id'].isin(links_small)]
```

```python
small_dataset.shape
```

```python
# Hàm literal_eval() sẽ đánh giá các giá trij trong cột 'cast' và trả về một danh sách các diễn viên. Tương tự với các cột còn lại

small_dataset['cast'] = small_dataset['cast'].apply(literal_eval)
```

```python
small_dataset['crew'] = small_dataset['crew'].apply(literal_eval)
```

```python
small_dataset['keywords'] = small_dataset['keywords'].apply(literal_eval)
```

```python
# Tạo cột 'cast_size' và 'crew_size'

small_dataset['cast_size'] = small_dataset['cast'].apply(lambda x: len(x))
small_dataset['crew_size'] = small_dataset['crew'].apply(lambda x: len(x))
```

```python
# Tạo cột 'Director'

def get_director(x):
    for i in x:
        if i['job'] == 'Director':
            return i['name']
    return np.nan
```

```python
small_dataset['director'] = small_dataset['crew'].apply(get_director)
```

```python
small_dataset['director']
```

```python
small_dataset['cast'][0]
```

```python
# Return top 3 actors
# Hàm lambda: check xem giá trị trong mỗi hàng của cột 'cast' có phải là list hay ko. Nếu là list thì return tên của tất cả 
#các diễn viên có trong list. 
# Hàm isinstance(): check xem mỗi hàng của cột 'cast' có phải là list. Nếu là list thì hàm lambda sẽ trả ra list tên actor, 
#nếu ko thì hàm lambda sẽ trả về list rỗng. 

small_dataset['cast'] = small_dataset['cast'].apply(lambda x: [i['name'] for i in x] if isinstance(x, list) else [])
```

```python
# Return top 3 actors
# Hàm lambda: check xem mỗi hàng có số diễn viên > 3 hay ko. Nếu > 3 => return 3 diễn viên đầu. Nếu <3 => trả về list gốc

small_dataset['top_actors'] = small_dataset['cast'].apply(lambda x: x[:3] if len(x) >= 3 else x)
```

```python
small_dataset['keywords'] = small_dataset['keywords'].apply(lambda x: [i['name'] for i in x] if isinstance(x, list) else [])
```
